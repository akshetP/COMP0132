```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
plume_center = np.array([5, 5])  # Center of the Gaussian plume
plume_sigma = 1.5  # Standard deviation of the Gaussian plume

# Create a grid of positions
x = np.linspace(0, 6, 100)  # Adjust the range as needed
y = np.linspace(0, 6, 100)  # Adjust the range as needed
X, Y = np.meshgrid(x, y)
positions = np.vstack([X.ravel(), Y.ravel()]).T

# Calculate Gaussian plume concentration at a given position
def calculate_concentration(position):
    distance = np.linalg.norm(position - plume_center)
    concentration = np.exp(-0.5 * (distance / plume_sigma) ** 2)
    return concentration

# Calculate concentrations for each position
concentrations = np.array([calculate_concentration(pos) for pos in positions])
concentrations = concentrations.reshape(X.shape)

# Create a contour plot
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, concentrations, cmap='plasma', levels=100)  # Adjust the colormap and levels as needed
plt.colorbar(label='Concentration')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Gaussian Plume Concentration')
plt.grid(True)
# Reverse the y-axis
plt.gca().invert_yaxis()
plt.show()
```
